```python
import cv2
from google.colab.patches import cv2_imshow # Import cv2_imshow for Colab

img = cv2.imread("sample3.jpg.png") # Corrected file name to include .png extension

# Always check if the image was loaded successfully
if img is None:
    print("Error: Could not read the image. Please ensure 'sample3.jpg.png' exists and the path is correct.")
else:
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    edges = cv2.Canny(gray, 100, 200)

    cv2_imshow(img) # Using cv2_imshow for displaying images in Colab
    cv2_imshow(edges) # Using cv2_imshow for displaying images in Colab

    # cv2.waitKey(0) and cv2.destroyAllWindows() are not typically used with cv2_imshow in Colab
```
