```python
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread("sample2.jpg.png") # Corrected file name to include .png extension

# Always check if the image was loaded successfully
if img is None:
    print("Error: Could not read the image. Please ensure 'sample2.jpg.png' exists and the path is correct.")
else:
    blur = cv2.GaussianBlur(img, (7, 7), 0)

    cv2_imshow(img)
    cv2_imshow(blur)
```
