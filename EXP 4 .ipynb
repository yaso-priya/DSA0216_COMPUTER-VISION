```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow # Import cv2_imshow for Colab

img = cv2.imread("sample4.jpg", 0)

# Check if image was loaded successfully
if img is None:
    print("Error: Could not read the image. Please ensure 'sample4.jpg' exists and the path is correct.")
    # If 'sample4.jpg' does not exist, you might want to try an existing image, e.g.:
    # img = cv2.imread("sample1.jpg", 0)
    # if img is None:
    #     print("Error: Could not read 'sample1.jpg' either.")
else:
    kernel = np.ones((5,5), np.uint8)
    dilate = cv2.dilate(img, kernel, iterations=1)

    cv2_imshow(img) # Using cv2_imshow for displaying images in Colab
    cv2_imshow(dilate) # Using cv2_imshow for displaying images in Colab

    # cv2.waitKey(0) and cv2.destroyAllWindows() are not typically used with cv2_imshow in Colab
```
